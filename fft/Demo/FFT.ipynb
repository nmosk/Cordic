```python
from pynq import Overlay
from pynq import allocate # replace with allocate for Pynq >= 2.7
import numpy as np

ol = Overlay('fft.bit')
fft_ip = ol.export_fft_0 # if you can't find the IP, type 'ol.' and hit Tab to see what options are available
```

```python
# set length and allocate arrays
length = 1024
inptR = allocate(shape=(length,), dtype=np.float32)
inptI = allocate(shape=(length,), dtype=np.float32)
outptR = allocate(shape=(length,), dtype=np.float32)
outptI = allocate(shape=(length,), dtype=np.float32)

# initialize input arrays with our input. Match to test bench
a = [i for i in range(length)]
b = [0 for i in range(length)]
np.copyto(inptR, a)
np.copyto(inptI, b)

fft_ip.write(0x30, length) #write to len
fft_ip.write(0x10, inptR.physical_address) #write to in_R from inpt
fft_ip.write(0x18, inptI.physical_address) #write to in_I from inpt
fft_ip.write(0x20, outptR.physical_address) #write to out_R from outp
fft_ip.write(0x28, outptI.physical_address) #write to out_R from outp
fft_ip.write(0x00, 1)

print("Real Output", "Imaginary Output \n")
for i in range(length):
    print(outptR[i], "\t\t  ", outptI[i])
```
